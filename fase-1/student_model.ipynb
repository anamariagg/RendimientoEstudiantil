# Predicción del rendimiento académico

Este modelo predice la nota de estudiantes según horas de estudio, inasistencias y apoyo familiar.

## Carga de librerías

En esta sección se importan las librerías necesarias para el análisis de datos y la construcción del modelo predictivo.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

## Carga de datos

Se carga el conjunto de datos que contiene información sobre estudiantes, incluyendo variables académicas y personales.

```python
data = pd.read_csv('students.csv')
data.head()
```

## Transformación de datos

Se convierten variables categóricas a numéricas para poder ser utilizadas en el modelo.

```python
data['apoyo_familiar'] = data['apoyo_familiar'].map({'si':1, 'no':0})
```

## Selección de variables

Se definen las variables independientes (X) y la variable objetivo (y), que corresponde a la nota final.

```python
X = data[['horas_estudio', 'inasistencias', 'apoyo_familiar']]
y = data['nota_final']
```

## Entrenamiento del modelo

Se entrena un modelo de tipo Random Forest para predecir el rendimiento académico de los estudiantes.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestRegressor(n_estimators=50)
model.fit(X_train, y_train)
```

## Predicción

Se realiza una predicción utilizando un ejemplo de estudiante.

```python
pred = model.predict([[3,2,1]])
print("Nota estimada:", pred)
```
